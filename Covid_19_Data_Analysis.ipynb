```python
# Importing necessary libraries

import pandas as pd          # Used for data manipulation and analysis
import numpy as np           # Used for numerical operations
import matplotlib.pyplot as plt   # Used for data visualization (graphs)
import seaborn as sns        # Used for advanced visualization
```

Step 6 – Data Cleaning


```python
df = df.dropna()
```

Step 7 – Convert Date Column

```python
df = pd.read_csv("/content/covid_19_clean_complete.csv")
df = df.dropna()

df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')
```

Explanation:-

Converts Date column into datetime format.

Useful for time analysis.

Step 8 – Total Cases by Country

```python
country_cases = df.groupby("Country/Region")["Confirmed"].sum()
country_cases.head()
```

Step 9 – Top 10 Affected Countries

```python
top_countries = df.groupby("Country/Region")["Confirmed"].sum().sort_values(ascending=False).head(10)
top_countries
```

Explanation:-

Sorts countries by highest cases.

Shows Top 10 affected countries.

Step 10 – Bar Chart Visualization

```python
plt.figure(figsize=(10,6))
top_countries.plot(kind='bar')
plt.title("Top 10 Countries with Highest COVID Cases")
plt.xlabel("Country")
plt.ylabel("Confirmed Cases")
plt.show()
```

Explanation

figure() → graph size

plot() → creates bar chart

title() → graph title

xlabel() and ylabel() → axis labels

Step 11 – COVID Trend Over Time

```python
daily_cases = df.groupby("Date")["Confirmed"].sum()

plt.figure(figsize=(12,6))
daily_cases.plot()
plt.title("Global COVID-19 Cases Over Time")
plt.xlabel("Date")
plt.ylabel("Confirmed Cases")
plt.show()
```

This graph shows:

📈 COVID spread trend across time

Step 12 – Death vs Recovery Analysis

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    x=df["Deaths"],
    y=df["Recovered"]
)

plt.title("Deaths vs Recovered Cases")
plt.show()
```

Scatter plot shows relationship between deaths and recoveries.

Step 13 – Correlation Heatmap

```python
plt.figure(figsize=(8,6))

numeric_df = df.select_dtypes(include=np.number)
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap")
plt.show()
```

Heatmap shows correlation between:

Confirmed

Deaths

Recovered
